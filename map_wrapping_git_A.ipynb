# Алгоритм

```python
def Range_dots(dots):
    print('dots', dots)
    l_dots = len(dots)                    # кол-во точек во входном массиве
    matrix = np.zeros((l_dots, l_dots))   # матрица расстояний
    visited = np.zeros(l_dots)            # индексы посещенных вершин в порядке обхода алгоритмом
    distances = np.zeros(l_dots)          # расстояния между вершинами в порядке обхода алгоритма

    # Если на вход даются два числа, то результат высчитывается отдельно
    if l_dots == 2:
        if (dots[0] >= 0 and dots[1] >= 0) or (dots[0] <= 0 and dots[1] <= 0):
            left  = min(dots[0], dots[1])
            right = max(dots[0], dots[1])
            distances = right - left
        else:
            neg = min(dots[0], dots[1])
            pos = max(dots[0], dots[1])
            dist_nl_pr = abs(neg) + pos
            dist_nr_pl = 180-pos + 180+neg
            if dist_nl_pr < dist_nr_pl:
                left  = neg
                right = pos
                distances = dist_nl_pr
            else:
                left  = pos
                right = neg
                distances = dist_nr_pl
    # Если на вход дается более чем 2 числа, то включается более сложная обработка
    if l_dots > 2:
        # Высчитывается матрица расстояний. Не симметричная матрица - верхний треугольник и нижний отличаются знаками
        for i in range(l_dots):
            for j in range(l_dots):
                matrix[i][j] = dots[i] - dots[j]
        
        i_curr = 0 # какая строка матрицы сейчас рассматривается. Порядок не обязательно восходящий - индексы скачут по оптимальным строчкам
        
        for i in range(l_dots): # проходимся по всем строкам матрицы
            min_p = 1000        # значение минимального положительного числа в текущей строке
            min_n = 1000        # значение минимального отрицательного числа в текущей строке
            num_n = 1           # Сколько отрицательных элементов
            j_select_n = -1     # индекс минимального отрицательного числа
            j_select_p = -1     # индекс минимального положительного числа
            distance = -1       # расстояние между выбранными числами
            for j in range(l_dots):                # проходимся по столбцам матрицы текущей строки 
                if matrix[i_curr][j] > 0:          # если работаем с положительными элементами, из них ищем минимум в строке
                    if matrix[i_curr][j] < min_p:
                        min_p = matrix[i_curr][j]
                        j_select_p = j             # запоминаем столбец, в котором находится текущий минимум
                if matrix[i_curr][j] < 0:          # если работаем с отрицательными элементами, из них ищем минимум в строке
                    num_n += 1                     # увеличиваем счетчик отрицательных чисел
                    if matrix[i_curr][j] < min_n:
                        min_n = matrix[i_curr][j]
                        j_select_n = j
            if num_n != l_dots: # Если в строке есть хоть 1 положительное число, то запоминаем минимальное из положительных
                visited[i]   = j_select_p
                distances[i] = min_p
                if distances[i] > 180:
                    distances[i] -= 360
                    distances[i] = abs(distances[i])
                i_curr = j_select_p
            else:
                visited[i]   = j_select_n
                distances[i] = abs(min_n)
                if distances[i] > 180:
                    distances[i] -= 360
                    distances[i] = abs(distances[i])
                i_curr = j_select_n
        
        max_dist = 0
        left  = -1000
        right = -1000
        l_index = 0
        r_index = 0
        for i in range(len(distances)):
            if distances[i] > max_dist:
                max_dist = distances[i]
                l_index = int(visited[i-1])
                left = dots[l_index]
                r_index = int(visited[i])
                right = dots[r_index]
    # Если на вход подается менее чем два числа, то возвращаюся границы -1000 
    elif l_dots < 2:
        left  = -1000
        right = -1000
        
    return left, right
```
